#### This dataset contains statistics on the world's billionaires, including information about their businesses, industries, and personal details. It provides insights into the wealth distribution, business sectors, and demographics of billionaires worldwide.
Key Features:

rank: The ranking of the billionaire in terms of wealth.

finalWorth: The final net worth of the billionaire in U.S. dollars.

category: The category or industry in which the billionaire's business operates.

personName: The full name of the billionaire.

age: The age of the billionaire.

country: The country in which the billionaire resides.

city: The city in which the billionaire resides.

source: The source of the billionaire's wealth.

industries: The industries associated with the billionaire's business interests.

countryOfCitizenship: The country of citizenship of the billionaire.

organization: The name of the organization or company associated with the billionaire.

selfMade: Indicates whether the billionaire is self-made (True/False).

status: "D" represents self-made billionaires (Founders/Entrepreneurs) and "U" indicates inherited or unearned wealth.

gender: The gender of the billionaire.

birthDate: The birthdate of the billionaire.

lastName: The last name of the billionaire.

firstName: The first name of the billionaire.

title: The title or honorific of the billionaire.

date: The date of data collection.

state: The state in which the billionaire resides.

residenceStateRegion: The region or state of residence of the billionaire.

birthYear: The birth year of the billionaire.

birthMonth: The birth month of the billionaire.

birthDay: The birth day of the billionaire.

cpi_country: Consumer Price Index (CPI) for the billionaire's country.

cpi_change_country: CPI change for the billionaire's country.

gdp_country: Gross Domestic Product (GDP) for the billionaire's country.

gross_tertiary_education_enrollment: Enrollment in tertiary education in the billionaire's country.

gross_primary_education_enrollment_country: Enrollment in primary education in the billionaire's country.

life_expectancy_country: Life expectancy in the billionaire's country.

tax_revenue_country_country: Tax revenue in the billionaire's country.

total_tax_rate_country: Total tax rate in the billionaire's country.

population_country: Population of the billionaire's country.

latitude_country: Latitude coordinate of the billionaire's country.

longitude_country: Longitude coordinate of the billionaire's country.

#### Droping columns with more than 70% null values 

```python
threshold = len(b) * 0.7  
b.dropna(thresh=threshold, axis=1, inplace=True)
```

#### Replacing null values 

```python
numerical_columns = b.select_dtypes(include=['number']).columns

# Replace null values in numerical columns with mean
b[numerical_columns] = b[numerical_columns].fillna(round(b[numerical_columns].mean()))

# Categorical columns
categorical_columns = b.select_dtypes(exclude=['number']).columns

# Replace null values in categorical columns with mode
for col in categorical_columns:
    mode_value = b[col].mode()[0]
    b[col].fillna(mode_value, inplace=True)
```

#### Outliers 

```python
# Select numerical columns
numerical_columns = b.select_dtypes(include=['int', 'float']).columns

for col in numerical_columns:
    plt.figure(figsize=(8, 6))
    b[col].plot(kind='box')
    plt.title(f'Box plot for {col}')
    plt.ylabel('Value')
    plt.show()
```

```python
outlier_threshold = 1.5  

numerical_columns = b.select_dtypes(include=['int', 'float']).columns

# Calculate outlier percentage for each numerical column
outlier_percentages = {}
for col in numerical_columns:
    q1 = b[col].quantile(0.25)
    q3 = b[col].quantile(0.75)
    iqr = q3 - q1
    
    lower_bound = q1 - outlier_threshold * iqr
    upper_bound = q3 + outlier_threshold * iqr
    
    outliers = b[(b[col] < lower_bound) | (b[col] > upper_bound)]
    outlier_percentage = len(outliers) / len(b) * 100
    outlier_percentages[col] = outlier_percentage

for col, percentage in outlier_percentages.items():
    print(f"Column '{col}': {percentage:.2f}% outliers")
```

### Data Analysis using Mysql 

#### 1. Find the total net worth of top 10  billionaires in the dataset.

```python
cur.execute('''  select personName,sum(finalWorth) AS NET_WORTH
                 from bs_data
                 group by personName
                 order by NET_WORTH desc
                 limit 10  
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 2. Find the Top 10 countries with most  billionaires .

```python
cur.execute('''  select country,count(personName)
                 from bs_data
                 group by country 
                 order by count(personName) desc
                 limit 10  
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 3. Calculate the total net worth of billionaires from each industry.

```python
cur.execute('''  select industries,sum(finalworth)
                 from bs_data
                 group by industries 
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 4. Find the youngest billionaire in each industry.

```python
cur.execute('''  WITH RankedBillionaires AS (
                    SELECT industries, personName, age,
                    RANK() OVER (PARTITION BY industries ORDER BY age) AS rnk
                    FROM bs_data
                 )
                 SELECT industries, personName
                 FROM RankedBillionaires
                 WHERE rnk = 1
                 group by industries
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 5. List the billionaires who are self-made and have a net worth above $50,000.

```python
cur.execute(''' select personName
                from bs_data
                where selfmade = "True"
                and finalworth > 50000
            ''' )
print(*cur.fetchall() , sep='\n')
```

#### 6. Find the top 3 countries with the highest average net worth of billionaires.

```python
cur.execute('''  select country,avg(finalWorth) as avg_worth
                 from bs_data
                 group by country
                 order by avg_worth desc
                 limit 3  
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 7. Count the billionaires whose birth year is between 1940 and 1950.

```python
cur.execute('''  select count(distinct personName)
                 from bs_data
                 where birthYear between 1940 and 1950  
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 8. Calculate the total net worth of billionaires born in the United States.

```python
cur.execute('''  select sum(finalWorth)
                 from bs_data
                 where country = "United States"  
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 9. Find the billionaire(s) with the highest net worth from each country.

```python
cur.execute(''' select country, personName, finalWorth
                from (
                    select country, personName, finalWorth,
                    row_number() over(partition by country order by finalWorth desc) as rn
                    from bs_data
                    ) as ranked
                WHERE rn = 1
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 10. Count the billionaires whose birth month is December and  net worth above 10000.

```python
cur.execute('''  select count(personName)
                 from bs_data
                 where birthMonth = 12
                 and finalWorth > 10000 
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 11. Calculate the difference in net worth between the youngest and oldest billionaires.

```python
cur.execute(''' with cte as(
                     select max(age) as mx_age ,min(age) as mi_age
                     from bs_data
                )
                select ((select finalworth from bs_data where age = (select mi_age from cte))
                      - (select finalworth from bs_data where age = (select mx_age from cte))) as net_diff;
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 12. Find the billionaire(s) whose source of wealth is in the technology industry and have a net worth above $70,000. 

```python
cur.execute('''  select personName
                 from bs_data
                 where industries='technology'
                 and finalWorth > 70000
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 13. Calculate the average age of male and female billionaires separately. 

```python
cur.execute('''  select gender,round(avg(age))
                 from bs_data
                 group by gender
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 14. List the billionaires who are not from the United States and have a net worth above $80,000.

```python
cur.execute('''  select distinct personName
                 from bs_data
                 where country != 'United States'
                 and finalworth > 80000
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 15. Count the billionaire(s) with the longest and shortest last names. 

```python
cur.execute('''  with c1 as (
                             select personName,length(lastName) as ln
                             from bs_data
                             group by personName
                             order by ln
                 )
                 select count(distinct personName) 
                 from bs_data
                 where length(lastName) = (select min(ln) from c1)
                 
                 union 
                 
                 select count(distinct personName) 
                 from bs_data
                 where length(lastName) = (select max(ln) from c1)
                 
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 16. Find the top 5 oldest billionaires.

```python
cur.execute('''  select personName,finalworth
                 from bs_data
                 order by age desc 
                 limit 5
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 17. Find the top 3 cities with the highest number of billionaires.

```python
cur.execute('''  select city,count(distinct personName) as pc
                 from bs_data
                 group by city
                 order by pc desc
                 limit 3
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 18. Find the billionaire(s) who have the same first name and last name.

```python
cur.execute('''  select distinct personName
                 from bs_data
                 where firstName = lastName
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 19. Count the billionaires whose last name starts with the letter 'A' and are from the United States.

```python
cur.execute('''  select distinct personName
                 from bs_data
                 where lastName like 'A%'
                 and country = 'United States'
            ''')
print(*cur.fetchall() , sep='\n')
```

#### 20. Calculate the average net worth of billionaires who are self-made versus those who inherited their wealth.

```python
cur.execute('''  with c2 as (
                             select status,avg(finalworth) as avg_worth
                             from bs_data
                             group by status
                 )
                 select * from c2
                 where status in ('U','D')
            ''')
print(*cur.fetchall() , sep='\n')
```
